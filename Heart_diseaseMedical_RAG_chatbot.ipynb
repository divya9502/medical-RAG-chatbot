```python
embeddings=SentenceTransformerEmbeddings(model_name="NeuML/pubmedbert-base-embeddings")
```

#Vector store creation

```python
vectorstore = Chroma.from_documents(chunks,embeddings)
```

```python
query=" what is heart disease"
search_results=vectorstore.similarity_search(query)
```

```python
search_results
```

```python
retriver = vectorstore.as_retriever(search_kwargs={"k": 5})
```

```python
retriver.get_relevant_documents(query)
```

##LLM Model

```python
llm=LlamaCpp(
    model_path="/content/drive/MyDrive/Biomistral/BioMistral-7B.Q4_K_M.gguf",
    temperature=0.2,
    max_tokens = 2048,
    top_p = 1
)
```

##use LLM and retriver and query, to generate final response

```python
template="""
<|context|>
you are a medical assistant that follows the instructions and generate the accurate responses based on the query and context provided.
please be truthful and give the correct answers only.
</s>
<|user|>
{query}
</s>
<|assistant|>
"""
```

```python
from langchain.schema.runnable import RunnablePassthrough
from langchain.schema.output_parser import StrOutputParser
from langchain.prompts import ChatPromptTemplate
```

```python
promt= ChatPromptTemplate.from_template(template)
```

```python
rag_chain = (
    {"context": retriver, "query": RunnablePassthrough()}
    | promt
    | llm
    | StrOutputParser()
)
```

```python
response = rag_chain.invoke(query)
```
